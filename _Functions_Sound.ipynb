```python
# Load Pictures function
def DicLoadSoundFiles(path):
    import csv
    import re
    import os
    
    File = []
    with open(path) as csvfile:
        inputfile = csv.reader(csvfile)
        for row in inputfile:
            File.append(row)
            
    # Retrieve pictures name
    pics = []
    for pos in File:
        test = ','.join(pos)
        # For some reason I have to put all those dots to account for 8-character words, will need to fix this later.
        pathname = os.path.splitext(path)[0] + '/(.|..|...|....|.....|......|.......|........).m4a'
        pic = re.search(pathname,test)
        if pic:
            found = pic.group(1)
            pics.append(found)
    return File, pics
            
# from pydub import AudioSegment

# [File,pics] = DicLoadSoundFiles('syllablesvoice.txt')

# # print(','.join(File[0]))

# count = 0
# for i in File:
#     test = ','.join(i)
#     song = AudioSegment.from_file(test,format="m4a")
#     newname = pics[count]+'.wav'
#     count = count+1
    
# #     song.export(newname,format='wav')
    
# print('Done')
```
